# explorasi data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("heart_attack_prediction_indonesia.csv")
```

### Distribusi Usia (Age)

```python
plt.figure(figsize=(8,5))
sns.histplot(df["age"], bins=20, kde=True)
plt.title("Distribusi Usia Pasien")
plt.xlabel("Usia")
plt.ylabel("Jumlah")
plt.show()
```

### Heart Attack Berdasarkan Gender

```python
plt.figure(figsize=(6,4))
sns.countplot(x="gender", hue="heart_attack", data=df)
plt.title("Perbandingan Heart Attack berdasarkan Gender")
plt.show()
```

### Heart Attack vs Smoking Status

```python
plt.figure(figsize=(8,5))
sns.countplot(x="smoking_status", hue="heart_attack", data=df)
plt.title("Heart Attack vs Smoking Status")
plt.xticks(rotation=45)
plt.show()
```

### Obesitas dan Heart Attack

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="heart_attack", y="obesity", data=df)
plt.title("Obesitas vs Heart Attack")
plt.show()
```

### Korelasi Faktor Risiko Utama

```python
plt.figure(figsize=(12,8))
sns.heatmap(df[["age","cholesterol_level","obesity","blood_pressure_systolic","blood_pressure_diastolic","triglycerides","heart_attack"]].corr(), annot=True, cmap="coolwarm")
plt.title("Korelasi Faktor Risiko dengan Heart Attack")
plt.show()
```
