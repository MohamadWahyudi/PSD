Pada tahap ini, data dieksplorasi untuk melihat nilai minimum dari setiap kolom numerik (Sepal Width, Sepal Length, Petal Width, dan Petal Length) berdasarkan kategori spesies bunga. Visualisasi menggunakan grafik batang di Power BI, sedangkan tabel menampilkan detail angkanya.


### Hasil Pengamatan:

- Iris-setosa
    - SepalWidthCm minimum: 2.30 cm
    - SepalLengthCm minimum: 4.30 cm
    - PetalWidthCm minimum: 0.10 cm
    - PetalLengthCm minimum: 1.00 cm
       
        → Terlihat bahwa setosa punya ukuran petal yang paling kecil dibandingkan spesies lain.

- Iris-virginica
    - SepalWidthCm minimum: 2.20 cm
    - SepalLengthCm minimum: 4.90 cm
    - PetalWidthCm minimum: 1.40 cm
    - PetalLengthCm minimum: 4.50 cm
           
        → Terlihat bahwa setosa punya ukuran petal yang paling kecil dibandingkan spesies lain.

- Iris-versicolor

    - SepalWidthCm minimum: 2.00 cm
    - SepalLengthCm minimum: 4.90 cm
    - PetalWidthCm minimum: 1.00 cm
    - PetalLengthCm minimum: 3.00 cm

        → Versicolor berada di tengah-tengah, nilai minimumnya lebih besar dari setosa tapi lebih kecil dibanding virginica.

Kesimpulan:

- Setosa punya ukuran petal terkecil → mudah dikenali dari ciri ini.

- Virginica punya petal terpanjang dan terlebar, bahkan nilai minimumnya sudah cukup besar.

- Versicolor posisinya ada di antara setosa dan virginica, baik dari segi panjang maupun lebar petal.

- max dari setiap kolom
image.png

Tahap berikutnya adalah melihat nilai maksimum dari setiap kolom numerik (Sepal Length, Sepal Width, Petal Length, dan Petal Width) berdasarkan spesies bunga. Data divisualisasikan dalam bentuk grafik batang dan tabel di Power BI.

### Hasil Pengamatan:

- Iris-setosa

    - PetalLengthCm maksimum: 1.90 cm
    - PetalWidthCm maksimum: 0.60 cm
    - SepalLengthCm maksimum: 5.80 cm
    - SepalWidthCm maksimum: 4.40 cm
        → Setosa terlihat punya petal paling pendek dan sempit, bahkan nilai maksimumnya jauh lebih kecil dibanding dua spesies lain.

- Iris-versicolor
    - PetalLengthCm maksimum: 5.10 cm
    - PetalWidthCm maksimum: 1.80 cm
    - SepalLengthCm maksimum: 7.00 cm
    - SepalWidthCm maksimum: 3.40 cm
        → Versicolor memiliki ukuran petal yang lebih besar daripada setosa, tapi masih lebih kecil dibanding virginica.

- Iris-virginica

    - PetalLengthCm maksimum: 6.90 cm
    - PetalWidthCm maksimum: 2.50 cm
    - SepalLengthCm maksimum: 7.90 cm
    - SepalWidthCm maksimum: 3.80 cm
        → Virginica mendominasi dengan ukuran terbesar, baik panjang petal maupun panjang sepal.

Kesimpulan:

- Setosa → mudah dikenali karena selalu punya petal kecil (bahkan nilai maksimalnya masih kecil).

- Versicolor → ukuran sedang, berada di antara setosa dan virginica.

- Virginica → jelas memiliki ukuran paling besar, sehingga perbedaan dengan dua spesies lain sangat terlihat.

2. rata rata setiap kolom
image.png

menghitung rata-rata dari setiap kolom numerik (Sepal Length, Sepal Width, Petal Length, dan Petal Width) berdasarkan spesies bunga. Visualisasi dilakukan dengan grafik batang dan tabel di Power BI.

### Hasil Pengamatan:
- Iris-setosa
    - Rata-rata PetalLengthCm: 1.46 cm
    - Rata-rata PetalWidthCm: 0.24 cm
    - Rata-rata SepalLengthCm: 5.01 cm
    - Rata-rata SepalWidthCm: 3.42 cm
        → Spesies ini memiliki petal paling kecil, namun sepal relatif lebih lebar dibanding dua spesies lainnya.

- Iris-versicolor
    - Rata-rata PetalLengthCm: 4.26 cm
    - Rata-rata PetalWidthCm: 1.33 cm
    - Rata-rata SepalLengthCm: 5.94 cm
    - Rata-rata SepalWidthCm: 2.77 cm
        → Versicolor berada di tengah-tengah, baik dari ukuran petal maupun sepal.

- Iris-virginica
    - Rata-rata PetalLengthCm: 5.55 cm
    - Rata-rata PetalWidthCm: 2.03 cm
    - Rata-rata SepalLengthCm: 6.59 cm
    - Rata-rata SepalWidthCm: 2.97 cm
        → Virginica mendominasi dengan rata-rata ukuran petal dan sepal yang paling besar.

Kesimpulan:
- Petal length & petal width menjadi ciri paling jelas untuk membedakan spesies.
- Setosa → petal rata-ratanya jauh lebih kecil, sedangkan sepalnya justru relatif lebih lebar.
- Versicolor → posisinya berada di tengah, cocok dijadikan batas antara setosa dan virginica.
- Virginica → punya rata-rata ukuran paling besar, sehingga paling mudah dikenali dari ukuran petalnya.


3. jumlah setiap kelas ditampilkan dalam grafik batang
image.png

langkah berikutnya adalah melihat jumlah data (count) untuk setiap spesies. Visualisasi berupa grafik batang memperlihatkan distribusi jumlah data pada setiap kelas bunga Iris.


### Hasil Pengamatan:

- Iris-setosa memiliki 50 baris data
- Iris-versicolor memiliki 50 baris data
- Iris-virginica memiliki 50 baris data

Jumlah ini konsisten di semua kolom numerik (Sepal Length, Sepal Width, Petal Length, Petal Width), masing-masing berisi 50 data untuk setiap spesies.

Kesimpulan:
- Dataset Iris terbagi merata: masing-masing spesies punya jumlah data yang sama yaitu 50 sampel.
- Hal ini penting karena dataset menjadi seimbang (balanced dataset) → tidak ada spesies yang mendominasi jumlah data.
- Kondisi ini ideal untuk pemodelan machine learning, karena tidak ada bias yang signifikan akibat ketidakseimbangan kelas.
